```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
df = pd.read_csv("result10000.csv", header=None)
df.head()
```

```python
index_to_link_dict = pickle.load(open("./index_to_link_dict.pickle","rb"))
```

```python
column_names = [index_to_link_dict[x] for x in range(len(df.columns))]
df.columns = column_names
```

```python
df.head()
```

```python
homepage = "https://www.hendrix.edu/"
links_which_do_not_link_to_homepage = []
for index, row in df.iterrows():
    if row[homepage] == 0:
        links_which_do_not_link_to_homepage.append(index)
```

```python
for link in links_which_do_not_link_to_homepage:
    print(index_to_link_dict[link])
```

```python
df.drop(df.columns[links_which_do_not_link_to_homepage],axis=1,inplace=True)
df.drop(df.index[links_which_do_not_link_to_homepage],inplace=True)
```

```python
for index, row in df.iterrows():
    if row[homepage] == 0:
        print(str(index) + " " + str(row))
```

```python
filtered_index_to_link = {}
filtered_link_to_index = {}

index = 0
for link in df.columns:
    filtered_index_to_link[index] = link
    filtered_link_to_index[link] = index
pickle.dump(filtered_index_to_link, open("filtered_index_to_link.p", "wb"))
pickle.dump(filtered_link_to_index, open("filtered_link_to_index.p", "wb"))
```

```python
df.to_csv("FilteredMatrix.csv",index=False)
```
